#### Dataset
The data we have has the percentage of women degrees in respective fields from 1970 to 2011 in USA. We'll have an interesting and insightful visualization with this data.

```python
import pandas as pd
import matplotlib.pyplot as plt
women_degrees = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
women_degrees
```

```python
plt.plot(women_degrees["Year"],women_degrees["Biology"])
plt.show()
```

From the plot, we can tell that Biology degrees increased steadily from 1970 and peaked in the early 2000's. We can also tell that the percentage has stayed above 50% since around 1987.

## Turn around

```python
plt.plot(women_degrees["Year"],women_degrees["Biology"], c="blue", label = "Women")
plt.plot(women_degrees["Year"],100 - women_degrees["Biology"], c="green", label = "Men")
plt.title("Percentage of Biology Degrees Awarded By Gender")
plt.legend(loc = "upper right")
plt.show()
```

A little more tweaking with the plot reveals that, in the first period, from 1970 to around 1987, women were a minority when it came to majoring in Biology while in the second period, from around 1987 to around 2012, women became a majority.

## Disabling tick marks & spines

```python
fig, ax = plt.subplots()
ax.plot(women_degrees["Year"],women_degrees["Biology"],c="blue",label="Women")
ax.plot(women_degrees["Year"],100-women_degrees["Biology"],c="green",label="Men")
ax.tick_params(left= False ,right= False,top= False,bottom=False)
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.set_title("Percentage of Biology Degrees Awarded By Gender")
ax.legend(loc="upper right")
plt.show()
```

## Comparing gender gaps in different streams
With improved data-ink ratio and use of colorblind friendly colors: orange color and dark blue color from the Color Blind 10 palette

```python
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
fig = plt.figure(figsize=(12, 12))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], c=(0/255,107/255,164/255), label='Women', linewidth = 3)
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[sp]], c=(255/255,128/255,14/255), label='Men', linewidth = 3)
    plt.ylim(0,100)
    plt.xlim(1968,2011)
    ax.tick_params(top=False,bottom=False,left=False,right=False)
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["left"].set_visible(False)
    plt.title(major_cats[sp])

# Calling pyplot.legend() here will add the legend to the last subplot that was created.
plt.legend(loc='upper right')
plt.show()
```

By looking at the plots, We can conclude that the gender gaps in "Computer Science" and "Engineering" are much bigger while the gaps in "Biology" and "Math and Statistics" are quite small. In addition, the first two degrees ("Computer Science" and "Engineering") categories are dominated by men whereas the latter degree categories are much more balanced. 

## Some more comparison 
With a more coherant and ordered layout of a single row with multiple columns and text annotations in place of legends

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    if sp == 0:
        ax.text(2005,87,"Men")
        ax.text(2002,8,"Women")
    if sp == 5:
        ax.text(2005,62,"Men")
        ax.text(2001,35,"Women")
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=(0/255,107/255,164/255), label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=(255/255,128/255,14/255), label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
plt.show()
```

The leftmost plot has the largest gender gap in 1968 while the psychology plot has the smallest gender gap in 1968. But here, we're more interested in the recent gender gaps in these Science, Technology, Engineering and Medical  degrees. Hence, it is ordered by decreasing ending gender gap, from the largest to smallest ending gender gaps.
